```python
import torch
import numpy as np
```

## Tensor の生成

```python
sample1 = torch.tensor([[1,2],[3,4]])
sample2 = torch.arange(0,10)
sample3 = torch.randn(100,10)
```

```python
print(f'sample1 = {sample1}')
print(f'sample2 = {sample2}')
print(f'sample3 = {sample3}')
```
